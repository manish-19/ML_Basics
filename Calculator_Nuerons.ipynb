```python
import numpy as np
import tensorflow as tf
```

```python
# Define the neural network architecture
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),  # Input layer with two numbers
    tf.keras.layers.Dense(64, activation='relu'),  # Hidden layer with 64 neurons
    tf.keras.layers.Dense(4, activation='softmax')  # Output layer with 4 neurons (one for each operation)
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Generate training data
# Inputs: pairs of numbers (x1, x2)
# Outputs: one-hot encoded labels for the operation (+, -, *, /)
x_train = np.array([[2, 3], [5, 1], [6, 6], [9, 3], [8, 4], [10, 2]])
y_train = np.array([[1, 0, 0, 0],  # Addition
                    [0, 1, 0, 0],  # Subtraction
                    [0, 0, 1, 0],  # Multiplication
                    [0, 0, 0, 1],  # Division
                    [1, 0, 0, 0],  # Addition
                    [0, 1, 0, 0]]) # Subtraction
```

```python
# Train the model
model.fit(x_train, y_train, epochs=100)

# Evaluate the model
loss, accuracy = model.evaluate(x_train, y_train)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

```python
# Function to get input values from user
def get_user_input():
    x1 = float(input("Enter first number: "))
    x2 = float(input("Enter second number: "))
    return x1, x2
```

```python
# Perform calculations using the trained model
def perform_calculation(x1, x2):
    input_data = np.array([[x1, x2]])
    prediction = model.predict(input_data)
    operation = np.argmax(prediction)
    if operation == 0:
        return x1 + x2
    elif operation == 1:
        return x1 - x2
    elif operation == 2:
        return x1 * x2
    elif operation == 3:
        if x2 == 0:
            return "Error: Division by zero"
        else:
            return x1 / x2

# Test the calculator
print("Performing calculations:")
x1, x2 = get_user_input()
print(f"{x1} + {x2} =", perform_calculation(x1, x2))
```
